```python
# STEP 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# STEP 2: Upload fraudTrain and fraudTest
uploaded = files.upload()

# STEP 3: Load CSVs
train_df = pd.read_csv("fraudTrain.csv")
test_df = pd.read_csv("fraudTest.csv")

# STEP 4: Drop 'Unnamed: 0' if exists
train_df.drop(columns=[col for col in ['Unnamed: 0'] if col in train_df.columns], inplace=True)
test_df.drop(columns=[col for col in ['Unnamed: 0'] if col in test_df.columns], inplace=True)

# STEP 5: Drop non-numeric columns
non_numeric_cols = train_df.select_dtypes(include=['object']).columns
train_df = train_df.drop(columns=non_numeric_cols)
test_df = test_df.drop(columns=non_numeric_cols)

# STEP 6: Define features and labels
X_train = train_df.drop("is_fraud", axis=1)
y_train = train_df["is_fraud"]

X_test = test_df.drop("is_fraud", axis=1)
y_test = test_df["is_fraud"]

# STEP 7: Train the Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)

# STEP 8: Make Predictions
y_pred = rf.predict(X_test)
y_prob = rf.predict_proba(X_test)[:, 1]

# STEP 9: Evaluation
print("\n Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\n Classification Report:")
print(classification_report(y_test, y_pred, digits=4))

roc_auc = roc_auc_score(y_test, y_prob)
print(f"\n ROC-AUC Score: {roc_auc:.4f}")
```
